```python
import pandas as pd
import censusdata
```

```python
census_data= censusdata.download('acs5', 2017, censusdata.censusgeo([('state', '34'), ('zip code', '*')]), ['B01001_001E', 'B19013_001E'])
census_pd = pd.DataFrame(census_data)

df = census_pd.rename(columns={"B01001_001E": "Population",
                                  
                                      "B19013_001E": "Household Income",
                              })
```

```python
df.reset_index(level=0, inplace=True)
```

```python

df1= df.rename(columns={"index": "place"}).astype(str)
df1.dtypes
```

```python
df1['place'] = df1['place'].str.split(',').str[0]
```
